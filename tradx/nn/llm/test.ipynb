```python
from mlx_lm import load, generate

model, tokenizer = load("mlx-community/Ministral-8B-Instruct-2410-4bit")

prompt="What was the Revenue in H1/24? Answer like this: {'H1/24': 12345}. Context: The revenue in 1H 2024 was €80.4 million."

if hasattr(tokenizer, "apply_chat_template") and tokenizer.chat_template is not None:
    messages = [{"role": "user", "content": prompt}]
    prompt = tokenizer.apply_chat_template(
        messages, tokenize=False, add_generation_prompt=True
    )

response = generate(model, tokenizer, prompt=prompt, verbose=True)
```

```python
p = "/Users/a2/code/fin/trade/test_docs/Financial Report INTRALOT S.A. (2024,Six-Month Statement,Both).pdf"
from pypdf import PdfReader

def extract_text_from_pdf(pdf_path):
    """
    Extrahiert Text aus einer PDF-Datei.
    
    :param pdf_path: Pfad zur PDF-Datei
    :return: Gesamter extrahierter Text als String
    """
    reader = PdfReader(pdf_path)
    extracted_text = ""
    
    # Iteriere durch alle Seiten der PDF
    for page in reader.pages:
        extracted_text += page.extract_text()
    
    return extracted_text



try:
    text = extract_text_from_pdf(p)
    print("Extrahierter Text:")
    print(text)
except Exception as e:
    print(f"Fehler beim Extrahieren des Texts: {e}")
```
