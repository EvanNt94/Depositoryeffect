```python
import pandas as pd

def get_data_around_stock_split(df):
    # Sicherstellen, dass der Index ein Datumsindex ist
    df.index = pd.to_datetime(df.index)
    
    # Den Tag mit dem Stock Split finden (wo Stock Splits ungleich 0 sind)
    split_dates = df[df['Stock Splits'] != 0].index
    
    # Falls kein Stock Split vorhanden ist, Rückmeldung geben
    if split_dates.empty:
        print("Kein Stock Split in den Daten gefunden.")
        return None
    
    # Annahme: Wir betrachten den ersten gefundenen Stock Split
    split_date = split_dates[0]
    
    # Zeitraum von ±5 Tagen um den Stock Split
    start_date = split_date - pd.Timedelta(days=5)
    end_date = split_date + pd.Timedelta(days=5)
    
    # Daten für diesen Zeitraum extrahieren
    data_around_split = df.loc[start_date:end_date]
    
    return data_around_split

# Funktion anwenden und das Ergebnis anzeigen
data_around_split = get_data_around_stock_split(df)
print(data_around_split)
```
