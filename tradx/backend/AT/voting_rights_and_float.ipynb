# erstmal in json format
# keys : float = 100-Σ(rights>5%), shareholders: [{name, total_rights}]


```python
import pandas as pd
from io import StringIO
from bs4 import BeautifulSoup
import requests, os, json
```

```python
link = "https://www.athexgroup.gr/web/guest/companies-info-voting-rights-"
soup = BeautifulSoup(requests.get(link).text, "lxml")
```

```python
id_main = "_votingrightsportlet_WAR_HelexServiceportlet_print-area"
div_base = soup.select("#"+ id_main)[0]
```

```python
names = div_base.select(".link-among-tables")
tables = div_base.select("table")
```

```python
def save_and_compute(name:str, df:pd.DataFrame):
    folder = "/Users/a2/code/fin/trade/data/fundamentals/company/free_float/AT"
    s =  '% Total Vot. Rights'
    ff = []
    for f in df.loc[:, s]:
        try:
            ff.append(float(f))
        except ValueError:
            continue
    free_float = 100 - sum(ff)
    free_float /= 100
    shareholders = []
    for n, sh in enumerate(df.loc[:, "Shareholder"]):
        diddy = {"name":sh, "total_rights":df.loc[n, s]}
        shareholders.append(diddy)
    with open(os.path.join(folder, name+".json"), "w") as f:
        json.dump({"float":free_float, "shareholders":shareholders}, f)


```

```python
for n, name_div in enumerate(names):
    name = name_div.select("div")[0].select("a")[1].text
    current_table = StringIO(tables[n].prettify())
    df = pd.read_html(current_table)[0]
    save_and_compute(name, df)
```
