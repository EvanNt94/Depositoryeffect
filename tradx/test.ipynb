```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from chronos import ChronosPipeline

os.environ["PYTORCH_ENABLE_MPS_FALLBACK"] = "1"
import torch
pipeline = ChronosPipeline.from_pretrained(
  "amazon/chronos-t5-large",
  device_map="cpu",
  torch_dtype=torch.bfloat16,
)
```

```python
X = torch.tensor([1 for _ in range(90)])

plen = 10
forecast = pipeline.predict(X, plen)
forecast.size()
```

```python
forecast_index = range(len(X), len(X)+plen)
low, median, high = np.quantile(forecast[0].numpy(), [0.1, 0.5, 0.9], axis=0)
plt.figure(figsize=(8, 4))
plt.plot(X, color="royalblue", label="historical data")
plt.plot(forecast_index, median, color="tomato", label="median forecast")
plt.fill_between(forecast_index, low, high, color="tomato", alpha=0.3, label="80% prediction interval")
plt.legend()
plt.grid()
plt.show()
```

```python
import re
def extract_hrefs(html_string):
    # Regex zum Extrahieren von href-Werten
    href_pattern = r'href=["\'](.*?)["\']'
    # Alle Ãœbereinstimmungen finden
    hrefs = re.findall(href_pattern, html_string)
    return hrefs

t = requests.get("https://www.athexgroup.gr/web/guest/rss-feeds/-/asset_publisher/hlxgrpannrss/content/id/7534197").text
extract_hrefs(t)
```
