```python
import numpy as np
import matplotlib.pyplot as plt
from py_vollib.black_scholes import black_scholes

S, K, T, r = 100, 105, 0.25, 0.01
C_market = 2.35
flag = 'c'

sigmas = np.linspace(0.01, 1.0, 200)
prices = [black_scholes(flag, S, K, T, r, sigma) - C_market for sigma in sigmas]

plt.plot(sigmas, prices, label="f(σ) = BS - C_market")
plt.axhline(0, color='black', linestyle='--')
plt.xlabel("σ (Volatilität)")
plt.ylabel("f(σ)")
plt.title("Zielfunktion für implizite Volatilität")
plt.legend()
plt.grid()
plt.tight_layout()
plt.savefig("iv_root_example.png")
```
